# Import Libraries & Load Data


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Google Play Store dataset
# FIX: Ensure the file path is correct. If running in Colab, upload the file first.
df = pd.read_csv("googleplaystore.csv")

print("Dataset loaded successfully!")
print(f"Shape: {df.shape}")
```

```python
df.head()

df.info()

df.describe()
```

# **DATA CLEANING**
1. Handle Missing Values

```python
# Check for missing values
print(df.isnull().sum())

# Drop rows where Rating is missing
df = df.dropna(subset=['Rating'])

# Clean 'Size' column
# Ensure 'Size' column is string type for string methods
df['Size'] = df['Size'].astype(str)

# Remove 'M', 'k', ',', '+' characters
df['Size'] = df['Size'].str.replace('M', '', regex=False)
df['Size'] = df['Size'].str.replace('k', '', regex=False)
df['Size'] = df['Size'].str.replace(',', '', regex=False)
df['Size'] = df['Size'].str.replace('+', '', regex=False)  # Use regex=False for literal '+'

# Replace 'Varies with device' with NaN
df['Size'] = df['Size'].replace('Varies with device', np.nan)

# Convert to numeric, coercing errors to NaN
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')

# Fill missing Size with median
df['Size'] = df['Size'].fillna(df['Size'].median())

print("Size column cleaned!")
```

2. Remove Duplicates

```python
df = df.drop_duplicates()

print(f"Duplicates removed! New shape: {df.shape}")
```

3. Clean "Installs" Column

```python
# Clean 'Installs' column
# FIX: Use .loc to avoid SettingWithCopyWarning
df.loc[:, 'Installs'] = df['Installs'].str.replace(',', '')
df.loc[:, 'Installs'] = df['Installs'].str.replace('+', '', regex=False)
df.loc[:, 'Installs'] = df['Installs'].str.replace('Free', '0', regex=False)  # Replace 'Free' with 0
df.loc[:, 'Installs'] = df['Installs'].astype(int)

print("Installs column cleaned!")
```

Clean "Price" Column

```python
# Clean 'Price' column
df.loc[:, 'Price'] = df['Price'].str.replace('$', '', regex=False)  # Remove '$'
df.loc[:, 'Price'] = df['Price'].str.replace('Free', '0', regex=False)  # Replace 'Free' with '0'
df.loc[:, 'Price'] = pd.to_numeric(df['Price'], errors='coerce')  # Convert to numeric, coercing errors to NaN

# Fill missing Price with 0 (or median if you prefer)
df['Price'] = df['Price'].fillna(0)

print("Price column cleaned!")
```

5. Clean "Size" Column (Note: Already processed above)

```python
# The 'Size' column has already been cleaned in an earlier step.
# This cell's content is redundant and can be skipped.
# No action needed here as the column is already processed.

print("Size column already cleaned. Moving on...")
```

6. Convert Date

```python
# Convert 'Last Updated' to datetime
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')

print("Date column converted!")
```

7. Check Final Cleaned Data

```python
print(df.info())
print("\n")
print(df.head())
print("\n")
print(df.describe())
```

# **EXPLORATORY DATA ANALYSIS (EDA)**

```python
# 1. Distribution of Ratings
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Rating', kde=True, bins=20)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

print(f"Average Rating: {df['Rating'].mean():.2f}")
print(f"Median Rating: {df['Rating'].median():.2f}")
```

```python
# 2. Top 10 Categories by Number of Apps
plt.figure(figsize=(12, 6))
top_categories = df['Category'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('Top 10 App Categories')
plt.xlabel('Number of Apps')
plt.ylabel('Category')
plt.show()

print(top_categories)
```

```python
# 3. Free vs Paid Apps
free_paid = df['Type'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=free_paid.index, y=free_paid.values, palette='Set1')
plt.title('Free vs Paid Apps')
plt.ylabel('Count')
plt.show()

print(free_paid)
print(f"\nPercentage of Free Apps: {(free_paid['Free'] / len(df) * 100):.2f}%")
```

```python
# 4. Relationship between Installs and Rating
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Installs', y='Rating', alpha=0.6, hue='Type')
plt.title('Relationship between Installs and Rating')
plt.xlabel('Installs')
plt.ylabel('Rating')
plt.xscale('log')
plt.show()
```

```python
# 5. Content Rating Distribution
plt.figure(figsize=(10, 6))
content_rating = df['Content Rating'].value_counts()
sns.barplot(x=content_rating.values, y=content_rating.index, palette='coolwarm')
plt.title('Apps by Content Rating')
plt.xlabel('Number of Apps')
plt.show()

print(content_rating)
```
